```python
#importing required libraries
import pandas as pd
import ast
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#loading the dataset
df=pd.read_csv(r"E:\tmdb_5000_credits.csv")
df
```

```python
df.dtypes
```

```python
#converting strings to list
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

df['cast'] = df['cast'].apply(convert)
df['crew'] = df['crew'].apply(convert)
```

```python
#combining cast and crew
df['tags'] = df['cast'] + df['crew']
```

```python
#converting list to string
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
#applying TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
vectors = tfidf.fit_transform(df['tags'])
```

```python
#Calculating cosine similarity
similarity = cosine_similarity(vectors)
```

```python
#Recommendation Function

def recommend(movie):

    index = df[df['title'].str.lower() == movie.lower()].index[0]

    scores = list(enumerate(similarity[index]))

    scores = sorted(scores, key=lambda x: x[1], reverse=True)[1:6]

    for i in scores:
        print(df.iloc[i[0]].title)
```

```python
#Testing
movie=input("Enter a movie name to recommend similar movies :")
recommend(movie)
```
